```python
import os
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nrep = 4
tmax = 600
t0 = 300
ti = []
lmbd = []
for i in range(nrep):
    exponent = i/(nrep - 1)
    ti.append(t0*(tmax/t0)**exponent)
    lmbd.append(t0/ti[-1])
```

```python
fig, ax = plt.subplots(4,1, sharex=True, sharey=True)
repl = np.loadtxt("replica_temp.xvg")
for i in range(nrep):
    ax[i].plot(repl[:,0], repl[:,i+1], '.', color='C%i'%i)
ax[-1].set_xlim(0,200)
ax[-1].set_xlabel('Time (ps)')
```

```python
nrep = 4
for i in range(nrep):
    command = "gmx_mpi energy -f rep%i/ener.edr -o rep%i/potential <<EOF\n10\n35\n36\n39\n40\n43\n44\nEOF"%(i,i)
    os.system(command)
```

```python
aver = []
fig, ax = plt.subplots()
for i in range(nrep):
    file = "rep%i/potential.xvg"%i
    data = np.loadtxt(file, comments=["#", "@"])
    ax.plot(data[:,0], data[:,1], label='$\lambda=$%g'%lmbd[i])
    aver.append(np.mean(data[:,1]))
ax.set_ylabel('Potential')
ax.set_xlim(0, data[-1,0])
ax.set_xlabel('Time (ps)')
plt.tight_layout()
```

```python
nbins = 25

fig, ax = plt.subplots()
for i in range(nrep):
    file = "rep%i/potential.xvg"%i
    data = np.loadtxt(file, comments=["#", "@"])
    hist, edges = np.histogram(data[20:,1], bins=np.linspace(-27250,-25750,nbins))
    centers = [np.mean(edges[i:i+2]) for i in range(nbins-1)]
    ax.plot(centers, hist)
ax.set_xlabel(r'$\langle V \rangle$ (kJ/mol)', fontsize=14)
plt.tight_layout()
```

```python
plt.plot(ti, aver, 'o')
plt.xlabel('$T_m$ (K)' )
plt.ylabel(r'$\langle V\rangle$ (kJ/mol)' )
```

```python
import matplotlib as mpl
import mdtraj as md
```

```python
fig, ax = plt.subplots(4,2, figsize=(10,5), sharex=True, sharey=True)
for i in range(4):
    traj = md.load_xtc("rep%i/traj_comp.xtc"%i, top="alaTB_ff03_tip3p_npt.gro")
    dih, phi = md.compute_phi(traj)
    dih, psi = md.compute_psi(traj)
    ax[i][0].plot(traj.time, 180./np.pi*phi.ravel(), '.', ms=1, color='C%i'%i)
    ax[i][1].plot(traj.time, 180./np.pi*psi.ravel(), '.', ms=1, color='C%i'%i)
ax[0][0].set_title('$\phi$')
ax[0][1].set_title('$\psi$')
ax[0][0].set_xlim(0,traj.time[-1])
plt.tight_layout()
```
